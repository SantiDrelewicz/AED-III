# Propiedades de los flujos en redes

## Ejercicio 1


Para cada una de las siguientes sentencias sobre el problema de flujo máximo en una red $N$: demostrar que es verdadera o dar un contraejemplo.

a) Si la capacidad de cada arista de $N$ es par, entonces el valor del flujo máximo es par.
> **Verdadero:** Sea $S$ un corte cualquiera de la red $N$
>
>\begin{equation}
⟹ F_{max} = \underset{S}{min}\{c(S)\} = \underset{S}{min} \sum_{ij\in S \bar{S}}
\underbrace{c(i → j)}_{\text{par}} ⟹ \text{F$_{max}$ es par}  \ \  \blacksquare
\end{equation}


b) Si la capacidad de cada arista de $N$ es par, entonces existe un flujo máximo en el cual el flujo sobre cada arista de $N$ es par.
> **Verdadero:** Lo demostramos por inducción en las iteraciones de ford-fulkerson. Miramos cuello botella en cada iteración:
> - **Caso Base:** inicialmente el flujo en la red es 0 y por lo tanto todas tienen sus capacidades disponibles que es par y su flujo es 0.
> - **Paso inductivo:** el flujo en cada arista es par. Si existe camino de aumento, el cuello botella será par. Al actualizar las capacidades de las aristas del camino de aumento, éstas son pares pues restamos flujo par a capacidad par. La clave aquí radica en el hecho de que, dado que todas las capacidades son pares, y está incrementando el flujo a través de un camino de aumento según la capacidad más pequeña restante (el cuello de botella), este incremento será par, ya que sería el mínimo de un conjunto de números pares (las capacidades restantes en el camino), y por lo tanto, también par.
>
> Este argumento se basa en el algoritmo de Ford-Fulkerson y cómo el proceso de búsqueda de caminos de aumento y actualización de flujos garantiza que, si empezamos con capacidades pares, cualquier flujo máximo encontrado respetará la condición de que cada flujo sobre cada arista sea par.

c) Si la capacidad de cada arista de $N$ es impar, entonces el valor del flujo máximo es impar.

d) Si la capacidad de cada arista de $N$ es impar, entonces existe un flujo máximo en el cual el flujo sobre cada arista de $N$ es impar.

>**Falso**
>
>image.png

>

> Lo importante es notar que la estructura de la red y la distribución del flujo pueden crear situaciones donde, incluso si todas las capacidades son impares, no necesariamente se requiere que cada flujo en el camino máximo sea impar para alcanzar el flujo máximo.
>
>Las propiedades del flujo máximo dependen de la estructura de la red y cómo las capacidades permiten diferentes configuraciones de flujo, más allá de las individualidades de cada arista.

e) Si todas las aristas de $N$ tienen capacidades racionales, entonces el flujo máximo es racional.

> **Verdadero:** Todas las capacidades de las aristas son racionales por hipótesis. Las operaciones realizadas para calcular el flujo (sumas y restas) no alteran la naturaleza racional de los números involucrados. Cualquier valor de flujo calculado, incluido el flujo máximo, conserva la propiedad de ser un número racional.

# Problemas de modelado I: caminos disjuntos en un grafo

## Ejercicio 5

Sea $G$ un digrafo con dos vértices $s$ y $t$.


a) Proponer un modelo de flujo para determinar la máxima cantidad de caminos disjuntos en aristas que van de $s$ a $t$.


> Tenemos el digrafo $G = (V, E)$ con $s,t \in V ⟹ N = (V, E)$ con función de capacidad: $c(e) = 1$ $∀$ $e∈E$
>
> ![Figura](figuras/fig65.PNG)

b) Dar una interpretación a cada unidad de flujo y cada restricción de capacidad.

> En este problema cada unidad de flujo representa un camino, como queremos pasar solo una vez por cada arista, la capacidad es $1$. Entonces el flujo
$F$ serían los caminos disjuntos que van desde $s$ a $t$.

c) Demostrar que el modelo es correcto

> Ahora queremos ver que si $F$ es el valor del flujo máximo en nuestra red
y $K$ es la cantidad de caminos disjuntos en aristas de nuestro
digrafo original $D$, entonces, $F = K$.
>
> **Qvq:** Existe un flujo válido en $N$ de valor $F ⇐⇒$ Existen $F$ caminos
disjuntos en aristas de $s$ a $t$ en $D$.
>
> Si probamos esta doble implicación entonces va a quedar
probado que $F = K$.
>
> **⟸)** sean $C_1, ..., C_F$ caminos disjuntos en aristas de $s$ a $t$ en D, definimos
>$$ f(e) = \begin{cases}
    1 & \text{si} \; e \in C_i, \; 1≤i≤F \\
    0 & c.c
\end{cases}$$
> Ahora qvq $f$ es un flujo válido y de valor F.
> - **Restricción de capacidad:** $f(e) ≤ c(e)$ $∀$ $e\in E(D)$ ✅ se cumple porque $f(e) ≤ 1 = c(e)$.
> - **Conservación del flujo:** $$\sum_{u∈N^-(v)}f(u→v) = \# \{i : v\in C_i, 1≤i≤F\} = \sum_{w∈N^+(v)}f(v→w)$$
> Esto se cumple ya que la # de aristas que entran y salen es igual a la # de caminos disjuntos a los cuales pertenece $v$.
> Como se cumplen estas dos restricciones $f$ es un flujo válido en $N$.
>
> - $f$ **es un flujo de valor** $F$: $$\sum_{v∈N^+(s)}f(s → v) = \#\{i : s \in C_i, 1≤i≤F\} = F$$
> puesto que $s \in C_i \; \forall \; 1≤i≤F$.
>
>**⟹)** Sea $f: E → \mathbb{N}$ un flujo valido de valor $F$ sobre $N$. Q.v.q existen F caminos disjuntos en aristas de $s$ a $t$.
>
> **Intución de la demo:**  la idea es ir construyendo los caminos disjuntos en aristas a partir del flujo $f$, todo esto a través de una demo inductiva.
>
> 1. $f$ es un flujo de valor $F$, y sabemos que $s$ y $t$ están conectados por las aristas $e$ con $f(e) = 1$; así que usando usando estas aristas tenemos un flujo que define un camino $P$ de $s$ a $t$
> 2. Luego, si **quitamos** estas aristas del flujo $f$,como $f(e) = 1 ⟹$ el valor del flujo $f' = f - P$ &nbsp; es &nbsp; $F' = F-1$. $f'$ sigue siendo un flujo válido y su valor es $F'=F-1$.
> 3. Luego, si volvemos a repetir este proceso sobre $f'$, el nuevo camino que obtendríamos, $P'$, no compartiría ninguna arista con $P$ puesto que $f(e) = 1$. Por lo tanto, si repetimos este proceso $F$ veces, vamos a lograr obtener $F$ **caminos disjuntos** en aristas. Así es cómo a partir del flujo $f$ puedo obtener $F$ caminos disjuntos en aristas de $D$ de $s$ a $t$.
>
>Así probamos la doble implicancia, por lo que queda demostrado que $F = K$ y que nuestro modelo representa correctamente el problema que queríamos resolver.

d) Determinar la complejidad de resolver el modelo resultante con el algoritmo de Edmonds y
Karp.

> Ya probamos que nuestro modelo es correcto. Por lo tanto, para encontrar la solución a nuestro problema alcanza con calcular el flujo máximo ⟹ utilizamos el algoritmo de Edmonds y Karp para Encontrar el flujo máximo.
>
> **Complejidad:** En este caso no es $O(n ∗ m^2
)$ porque la complejidad está dada por $min(O(n ∗ m^2), O(F ∗ m))$, y en este caso $F ≤ n$ puesto que la cantidad de caminos disjuntos en aristas no puede ser mayor a n ⟹ $T_{EK} = O(n * m)$

## Ejercicio 7 **(Matching de cardinalidad máxima en grafos bipartitos)**

Nos dan una lista de personas y sus preferencias para hacer ciertas tareas. Cada persona puede hacer una tarea y además cada tarea solo puede ser hecha por una sola persona. Nos piden calcular la cantidad máxima de tareas que se pueden hacer.

a) Modelar el ejercicio como un problema de flujo máximo.

> Resumiendo el problema:
>- Tenemos un conjunto de personas $P$ y un conjunto de tareas $T$.
>- Cada persona puede realizar algunas tareas (no necesariamente todas).
>- Cada persona puede realizar como máximo una tarea.
>- Cada tarea puede ser realizada por una sola persona.
>- Queremos maximizar la cantidad total de tareas realizadas.
>
> Queremos modelar este problema como un flujo máximo.
>
> ![Figura](figuras/fig671.PNG)
>
> Agregamos los nodos $s$ y $t$ para poder construir una red de flujo
dirigida.
>
> ![Figura](figuras/fig672.PNG)
>
> Ahora definimos los arcos y sus capacidades:
>- Desde $s$ hacia cada tarea $t ∈ T$, un arco de capacidad $1$.
>- Desde cada tarea $t$ hacia las personas $p$ que puede hacer, arcos de capacidad $1$.
>- Desde cada persona $p ∈ P$ hacia el sumidero t, también arcos de capacidad $1$.
>
> La red de flujo resultante tiene la siguiente estructura por capas: $ s → T → P → t$
>
> ![Figura](figuras/fig673.PNG)
>
>**Semántica de la unidad de flujo:** Una unidad de flujo representa una persona asignada a una tarea. Entonces, un flujo válido de valor *k* equivale a un matching de tamaño *k*.


b) Mostrar que el modelado es correcto.
> Queremos probar que:
“Existe un matching de tamaño k” ⇔ “Existe un flujo válido de valor k”
>
> $\textbf{⟹)}$ Si existe un matching de tamaño k, construimos el siguiente flujo:
> - Para c/par $(t, p)$ incluido en el matching, envíamos una unidad de flujo por el camino $$ s → t → p → t $$ este flujo respeta las capacidades (todas son $1$) y conserva flujo en todos los vértices. El valor total del flujo es *k*.
>
> $\textbf{⟸)}$ Si existe un flujo válido de valor *k*, como las capacidades son enteras, por el teorema de integridad del flujo máximo, el flujo puede considerarse entero.
>
> Por lo tanto, c/arco $t → p$ con flujo $1$ indica que la persona $p$ realiza la tarea $t$. Como las capacidades son $1$, ningún vértice está involucrado en más de una asignación.
>
> El conjunto de arcos con flujo $1$ define un matching de tamaño *k*.

c) Dar una cota superior de la complejidad
> **Acotamos nodos, arcos y flujo máximo:**
>- Nodos: $|P| + |T| + 2 = \mathcal{O}(|P| + |T|)$
>- Arcos:
>   - |P| arcos desde s,
>   - A lo sumo |P||T| arcos entre $P$ y $T$.
>   - |T| arcos hacia t,
>
>   Total: $\mathcal{O}(|P||T|)$.
> **Cota de flujo máximo:** por el teorema de corte mínimo, el valor del flujo máximo es igual al tamaño del corte mínimo
>- Hay un corte que separa a $s$ de todos los nodos $T$, de tamaño $|T|$.
>- Otro corte separa a todos los nodos $P$ de $t$, de tamaño $|P|$ $$⟹ F_{max} ≤ mín(|P|, |T|).$$
>
> **Complejidad del algoritmo:**
>- Sabemos que $n = \mathcal{O}(|P| + |T|)$ y $m = \mathcal{O}(|P||T|)$.
>- Por el algoritmo de Edmonds-Karp, la complejidad es: $$\mathcal{O}(min(mF, nm^2))$$ donde $F$ es el valor del flujo máximo.
>
> Reemplazando por las cotas:
> $$\mathcal{O}(|P||T| · mín(|P|, |T|)).$$

## Ejercicio 9

Sean $r_1, . . . , r_m$ y $c_1, . . . , c_n$ números naturales. Se quiere asignar los valores de las celdas de una matriz de $m × n$ con números naturales de forma tal que la $i$-ésima fila sume $r_i$ y la i-ésima columna sume $c_i$.

a) Modelar el problema de asignación como un problema de flujo.

> Pensamos en el siguiente grafo: el tablero nos define dos conjuntos bipartitos, el valor de una columna $ c_i $ solo depende del valor de las filas $ f_1, f_2, \ldots, f_n $ en la posición de la columna. Eso nos define una relación entre las particiones. Podemos modelar estas relaciones como un grafo, teniendo a cada $ c_i $ y $ f_j $ como vértices en él, siendo las aristas las relaciones correspondientes. Observemos que cada columna se relaciona con todas las filas y viceversa. Luego podemos establecer nuestra fuente $ s $ y nuestro sumidero $ t $. Las capacidades entre la fuente y las columnas son iguales a $ c_i $ para cada columna, y las de las filas al sumidero son $ r_j $. Las capacidades de las aristas intermedias son infinitas (ya están acotadas por las capacidades que salen de la fuente).
>
> image.png

b) Dar una interpretación a cada unidad de flujo y cada restricción de capacidad.

> Cada unidad de flujo equivale a sumarle una unidad a esa celda en la matriz.
>- Las restricciones de capacidad $r_i$ y $c_i$ $∀$ $i \in \{1, ...m\}$, $j \in \{1,...,n\}$ indican cuánto deben sumar la fila $r_i$ y la columna $c_j$ respectivamente.
>- Las restricciones de capacidad $∞$ indican que no hay restricción para el valor que puede tener una celda individual.

c) Demostrar que el modelo es correcto.

> **Queremos ver que**
$$∃ \text{ "asignación posible" para } A \in \mathbb{N}^{m × n}$$ $$⟺$$ $$∃ f \text{ función de flujo válido en } N = (V, E) \text{ tal que } F = \sum_{i=1}^m r_i$$
donde "asignación" es la matriz de $m × n$ y en cada celda hay un número natural y "posible" es:
$$(1)\sum_{j=1}^n A_{ij} = r_i \ ∀ \ i \in \{1,...,m\}$$
$$(2)\sum_{i=1}^m A_{ij} = c_j \ ∀ \ j \in \{1,...,n\}$$
>
> $\textbf{⟸)}$ Sea $f$ una función de flujo máximo en el modelo tal que $F = \sum_{i=1}^m r_i$. Construyo la "asignación posible" $A_{ij} = f(f_i → c_j)$
>- Veamos que es asignación: $i \in \{1,...,m\}$ y $j \in \{1,...,n\} ⟹$ hay $m × n$ pares $(f_i,c_j) ⟹ A$ tendrá dimensión $m × n$ y $A_{ij} \in \mathbb{N}$.
>- Veamos que es posible
>   1. Como $F = ∑_{i=1}^m r_i$ ⟹ cada arista $(s, f_i)$ está saturada porque estoy sumando el total de su capacidad al flujo ⟹ por conservación de flujo: $$\text{" flujo que entra al nodo" } f_i \ ∀i \in \{1,...,m\} \ r_i = ∑_{j=1}^n f(f_i → c_j) =\text{ "flujo que sale del nodo" } r_i = ∑_{j=1}^n A_{ij}$$ "Aplico la asignación que hice" ⟹ se cumple (1)
>   2. Con el mismo argumento de arriba pero ahora con los nodos $c_j$ concluimos que se cumple (2)
>
>Luego existe "asignación posible" para $A ∈ \mathbb{N}^{m × n}$
>
> $\textbf{⟹)}$ Sea $A ∈ \mathbb{N}^{m × n}$ una matriz que cumple "asignación posible". Construyo la función de flujo máximo $
f: E → \mathbb{N}$ de valor $F = \sum_{i=1}^m r_i$
>
>\begin{equation}
f(u, v) = \begin{cases}
r_i & \text{si} \; u = s ∧ v = r_i, \ i = \{1,...,m\} \\
c_j & \text{si} \; u = c_j ∧ v = t, \ j = \{1,...,n\} \\
A_{uv} & \text{si no}
\end{cases}
\end{equation}
>
> Veamos que $f$ es un flujo válido
>- **Restricciones de capacidad:** lo cumple porque las únicas aristas con capacidad finita son las $(s, f_i)$ y $(c_j, t)$ y en todas ellas el flujo es igual a la capacidad ⟹ es $≤$ a la capacidad.
>
>- **Conservación de flujo:**
>
>   - Quiero ver que los nodos $f_i$ conserva el flujo: o sea, $f(s, f_i) = \sum_{j=1} f(f_i, c_j)$. Comienzamos con $f(s, f_i) = r_i$ por la definición de $f$ y por "asignación posible" $r_i = ∑_{j=1}^n A_{ij}$ que a su vez por definición de $f$, $∑_{j=1}^n A_{ij} = ∑_{j=1}^n f(f_i, c_j)$ ⟹ Los nodos $r_i$ conservan el flujo.
>
>   - Queremos ver que los nodos $c_j$ conservan el flujo: siguiendo la idea de arriba concluimos que los nodos $c_j$ conservan el flujo.
>
>   Como todos los nodos distintos a $s$ y $t$ conservan flujo ⟹ f conserva flujo ✅
>
>- Veamos que $F = \sum_{i=1}^m r_i$: comenzamos por la definición de flujo ⟹ $F$ es igual a la cantidad de flujo que sale del nodo $s$ ⟹ $F = \sum_{i=1} f(s, f_i) = \sum_{i=1}^m r_i$ por definición de $f$.
>- Veamos que $f$ es máximo: como $f(s, f_i) = r_i$ y la capacidad de la arista $(s, f_i)$ es $r_i$ $∀i∈\{1,...,m\}$ ⟹ cada arista $(s, f_i)$ está saturada y no es posible enviar más flujo que el que se está enviando porque todas las aristas salientes de s están saturadas ⟹ $f$ es de flujo máximo.
>- Luego existe una función $f: E → \mathbb{N}$ función de flujo máximo en N = (V, E, c) tal que $F = \sum_{i=1}^m r_i$ ✅

d) Determinar la complejidad de resolver el modelo resultante con el algoritmo de Edmonds y Karp.

>- $|E| = m + mn + n = \mathcal{O}(nm)$
> - Complejidad: $\mathcal{O}(|V||E|^2) = \mathcal{O}((n+m)n^2m^2)$
